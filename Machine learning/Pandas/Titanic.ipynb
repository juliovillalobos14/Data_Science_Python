# Carga de datos a travez de la función open

```python
data1 = open("titanic3.csv",'r')
```

```python
cols = data1.readline().strip().split(",")
cols
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
main_dict
```

```python
for line in data1:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
    
print("El data set tiene %d filas"%(counter))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

## Carga de datos desde un fichero

```python
fichero = open("Customer Churn Model.txt",'r')
```

```python
clm = fichero.readline().strip().split(",")
clm
```

```python
n_cols = len(clm)
n_cols
```

```python
counter = 0
main_dict = {}
for cl in clm:
    main_dict[cl] = []
    
main_dict
```

```python
for line in fichero:
    values = line.strip().split(",")
    for i in range(len(clm)):
        main_dict[clm[i]].append(values[i])
    counter += 1
    
print("El data set tiene %d filas y %d columnas"%(counter, n_cols))
```

```python
df1 = pd.DataFrame(main_dict)
df1.head()
```

## Lectura y escritura de ficheros 

```python
infile = "Customer Churn Model.txt"
outfile = "Tab Customer Churn Model.txt"
```

```python
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4 = pd.read_csv(outfile, sep = "\t")
df4.head()
```
