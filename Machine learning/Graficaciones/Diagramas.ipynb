```python
#Graficaciones
#Importación de librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Lectura de los datos con pandas
datos = pd.read_csv('Movies20.csv')
#Creación de dataframe
df = pd.DataFrame(datos)
```

```python
#Agrupamiento por generos de acuerdo al Metascore
#Realizará la suma y graficará en barras
df.groupby('Genre')['Metascore'].sum().plot(kind='bar',legend='Reverse')
```

```python
#Agrupamiento por generos de acuerdo al Metascore
#Realizará la suma y graficará en barras
df.groupby('Genre')['Metascore'].sum().plot(kind='barh',legend='Reverse')
plt.xlabel('Suma de Score')
```

```python
df.Metascore.groupby(df.Genre).sum().plot(kind = 'pie',cmap = 'Paired')
plt.axis('equal')
```

```python
df.Metascore.groupby(df.Genre).sum().plot(kind = 'pie',cmap = 'Paired')
plt.ylabel('equal')
```

```python
#cmap --> Son los colores que se utilizarán
df.Metascore.groupby(df.Genre).mean().plot(kind = 'pie',cmap = 'Paired')
plt.title('Promedio de score')
```

```python
#DIAGRAMA DE DISPERSIÓN
# lmlplot --> hace correlaciones de tipo lineal
#hue --> colores basados en el rango, entre mayor más oscuro
#palette --> Paleta de colores
sns.lmplot(x = 'Votes', y = 'Metascore', data = df, fit_reg = False, hue = 'Votes', legend = False, palette = 'Paired', aspect = 2)
```

```python
#TABLA PIVOTE
#Relación entre varias columnas
nuevo = df[['Genre','Year','Votes','Metascore']]
sns.set(style = 'ticks', color_codes = True)
g = sns.pairplot(nuevo, hue = 'Genre', palette = 'Spectral')
```

```python
d = sns.lmplot(x = 'Votes', y = 'Metascore', data = df, palette = 'Set1')
```

```python
r = sns.lmplot(x = 'Votes', y = 'Metascore', data = df, hue = 'Genre', col = 'Year', aspect = .4, x_jitter = .1)
```

```python
t = sns.lmplot(x = 'Votes', y = 'Metascore', data = df, hue = 'Genre', palette = 'Set1')
```
