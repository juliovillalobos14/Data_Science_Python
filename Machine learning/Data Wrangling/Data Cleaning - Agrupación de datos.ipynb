# Agregación de datos por categoria

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gender  = ['Male', 'Female']
income = ['Poor', 'Middle Class', 'Rich']
```

```python
#Tamano del dataset de 500 filas
n = 500

gender_data = []
income_data = []

for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
income_data[1:8]
```

```python
#Generar datos númericos
# z -> N(0,1)
#N(m, s) - > m + s + z
height = 160 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
#Creación del dataframe
data = pd.DataFrame(
    {
        'Gender': gender_data,
        'Economic Status': income_data,
        'Height': height,
        'Weight': weight,
        'Age': age,
        'Income': income
    }
)
```

```python
data.head(12)
```

```python
#Conversión de los datos a números enteros
```

### Agrupación de datos por categoría 

```python
groupby_gender = data.groupby('Gender')
```

```python
groupby_gender.groups
```

```python
for names, groups in groupby_gender:
    print(names)
    print(groups)
```

```python
groupby_gender.get_group('Female')
```

```python
#Agrupación de más de una categoría
groupby_double = data.groupby(['Gender','Economic Status'])
```

```python
len(groupby_double)
```

```python
for names, groups in groupby_double:
    print(names)
    print(groups)
```

### Operaciones sobre datos agrupados

```python
groupby_double.sum()
```

```python
groupby_double.mean()
```

```python
groupby_double.size()
```

```python
groupby_double.describe()
```

```python
groupby_income = groupby_double['Income']
```

```python
groupby_income.describe()
```

```python
groupby_double.aggregate(
    {
        'Income': np.sum,
        'Age': np.mean,
        'Height': np.std
    }
)
```

```python
groupby_double.aggregate(
    {
        'Age': np.mean,
        'Height': lambda h:(np.mean(h))/np.std(h)
    }
)
```

```python
groupby_double.aggregate((np.sum, np.mean, np.std))
```

```python
#Agregar con lambda
groupby_double.aggregate([lambda x: np.mean(x) / np.std(x)])
```

## Filtrado de datos 

```python
varfiltrado = groupby_double['Age'].filter(lambda x: x.sum() > 2400)
```

```python
varfiltrado
```

## Transformación de variables 

```python
zscore = lambda x: (x - x.mean()) / x.std()
```

```python
zgroup = groupby_double.transform(zscore)
```

```python
plt.hist(zgroup['Age'])
```

```python
fill_na_mean = lambda x: x.fillna(x.mean)
```

```python
groupby_double.transform(fill_na_mean)
```
